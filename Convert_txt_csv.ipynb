```
from google.colab import drive
import pandas as pd
import os
drive.mount('/content/drive')
import numpy as np
```

```
DATA_DIR = 'drive/My Drive/Self Driving'
TXT_FILE = 'T_LOG.TXT'
FILE_PATH = os.path.join(DATA_DIR,TXT_FILE)

with open(FILE_PATH, 'r') as file:
    data = file.read()

print(data)
```

```
#Splitting TXT
dataSplit = data.split('\n')
#removing blank lines
while("" in dataSplit) : 
  dataSplit.remove("")
#removing first line
dataSplit.pop(0)

for each_element in dataSplit:
  if each_element[0] == ">":
    dataSplit.remove(each_element)
#printing contents of the array
for elements in dataSplit:
  print(elements)
```

```
#Creating Training and Testing Vectors:
X_TRAIN = []
Y_TRAIN = []

X_sample = []
for i in range(len(dataSplit)):
  if((i+1)%4==0):
    reading = dataSplit[i].split(':')
    temp = reading[1]
    actual_reading = temp[1:-1]
    final_reading = actual_reading.split(" ")
    Y_value = 0
    if '"LEFT"' in final_reading:
      Y_value = -1
    if '"RIGHT"' in final_reading:
      Y_value = 1
    Y_TRAIN.append(Y_value)
    X_TRAIN.append(X_sample)
    X_sample=[]
  else:
    reading = dataSplit[i].split(':')
    actual_reading = reading[1]
    final_reading = actual_reading[1:-3]
    X_sample.append(int(final_reading))

numpy_X = [np.asarray(k) for k in X_TRAIN]
NPX = np.asarray(numpy_X)

numpy_Y = []
for j in Y_TRAIN:
  if j==-1:
    subArr=[1,0,0]
  elif j==0:
    subArr=[0,1,0]
  else:
    subArr=[0,0,1]
  tempArr = np.asarray(subArr)
  numpy_Y.append(tempArr)
NPY = np.asarray(numpy_Y)


print(NPX)
print(NPY)
print(len(NPX))
print(len(NPY))
```

```
# saving NPX and NPY numpy arrays
X_LOC = 'drive/My Drive/Self Driving/XTRAIN.npy'
Y_LOC = 'drive/My Drive/Self Driving/YTRAIN.npy'

np.save(X_LOC, NPX)
np.save(Y_LOC, NPY)
```
