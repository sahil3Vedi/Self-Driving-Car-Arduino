```
from google.colab import drive
import pandas as pd
import os
drive.mount('/content/drive')
import numpy as np
```

```
X_LOC = 'drive/My Drive/Self Driving/XTRAIN.npy'
Y_LOC = 'drive/My Drive/Self Driving/YTRAIN.npy'

X_TRAIN = np.load(X_LOC)
Y_TRAIN = np.load(Y_LOC)
```

```
import os
import numpy as np
from sklearn.svm import LinearSVC
import random
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.utils import to_categorical

```

```
def preprocess(x, y):
  x = tf.cast(x, tf.float32) / 255.0
  y = tf.cast(y, tf.int64)

  return x, y

def create_dataset(xs, ys, n_classes=10):
  ys = tf.one_hot(ys, depth=n_classes)
  return tf.data.Dataset.from_tensor_slices((xs, ys)) \
    .map(preprocess) \
    .shuffle(len(ys)) \
    .batch(500)
```

```
model2 = tf.keras.Sequential([
# Adds a densely-connected layer with 64 units to the model:
layers.Dense(10, activation='relu', input_shape=(3,)),
# Add another:
layers.Dense(15, activation='relu'),
# Add an output layer with 10 output units:
layers.Dense(3, activation='softmax')])

model2.compile(optimizer=tf.keras.optimizers.Adam(0.1),
              loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model2.fit(X_TRAIN, Y_TRAIN, epochs=100, batch_size=32)
```

```
model2.summary()
```

```
weights = []
biases = []

for i in range(len(model2.layers)):
  weightMx = model2.layers[i].get_weights()
  weights.append(weightMx[0])
  biases.append(weightMx[1])

rounded_weights = []
rounded_biases = []

for j in weights:
  Wtemp = []
  for x in j:
    temp = [round(n,1) for n in x]
    Wtemp.append(temp)
  rounded_weights.append(Wtemp)

for k in biases:
  temp = [round(n,1) for n in k]
  rounded_biases.append(temp)

print(rounded_weights)
print(rounded_biases)
```
